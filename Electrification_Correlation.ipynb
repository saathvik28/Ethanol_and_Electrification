## Calculating ratios

### Module(s) Used: pandas
Now, we shall calculate the relative EV fuel efficiency ratios for both small SUVs and midsize sedans.
These ratios are relative to the average Miles Per Gallon(MPG) value of regular gas vehicles of the vehicle classes.

I used specific Lexus vehicles(ES and NX) for calculating the relative ratios as they are one of the only vehicles of midsize car and small SUV varieties respectively that have key similarities between its hybrid as well as gas variants in the dataset. This makes it crucial for comparing fuel efficiencies among hybrid, gasoline and other degrees of electrification.

```python
city_midsize = ev_midsize['City MPG']
city_midsize = city_midsize.astype(int)
city_midsize1 = city_midsize.mean()
city_midsize = city_midsize1 / int(es250_city.values[0])

hwy_midsize = ev_midsize['Hwy MPG']
hwy_midsize = hwy_midsize.astype(int)
hwy_midsize = hwy_midsize.mean() / int(es250_hwy.values[0])

cmb_midsize = ev_midsize['Cmb MPG']
cmb_midsize = cmb_midsize.astype(int)
cmb_midsize = cmb_midsize.mean() / int(es250_cmb.values[0])

ev_midsize_ratios = (city_midsize, hwy_midsize, cmb_midsize)


city_smallsuv = ev_smallsuv['City MPG']
city_smallsuv = city_smallsuv.astype(int)
city_smallsuv = city_smallsuv.mean() / int(nx250_city.values[0])

hwy_smallsuv = ev_smallsuv['Hwy MPG']
hwy_smallsuv = hwy_smallsuv.astype(int)
hwy_smallsuv = hwy_smallsuv.mean() / int(nx250_city.values[0])

cmb_smallsuv = ev_smallsuv['Cmb MPG']
cmb_smallsuv = cmb_smallsuv.astype(int)
cmb_smallsuv = cmb_smallsuv.mean() / int(nx250_city.values[0])

ev_smallsuv_ratios = (city_smallsuv, hwy_smallsuv, cmb_smallsuv)
```

## Calculating ratios

### Module(s) Used: pandas

Now, we shall move on to calculating the relative PHEV fuel efficiency ratios for both small SUVs and midsize sedans.
These ratios are relative to the average Miles Per Gallon(MPG) value of regular gas vehicles of the vehicle classes.

```python
city_midsize = pd.DataFrame()
city_midsize[['Gas', 'Hybrid']] = phev_midsize['City MPG'].str.split('/', expand=True)
city_gas_midsize = city_midsize['Gas'].astype(int).mean()
city_ph_midsize = city_midsize['Hybrid'].astype(int).mean()

city_phev_midsize = city_ph_midsize / city_gas_midsize


hwy_midsize = pd.DataFrame()
hwy_midsize[['Gas', 'Hybrid']] = phev_midsize['Hwy MPG'].str.split('/', expand=True)
hwy_gas_midsize = hwy_midsize['Gas'].astype(int).mean()
hwy_ph_midsize = hwy_midsize['Hybrid'].astype(int).mean()

hwy_phev_midsize = hwy_ph_midsize / hwy_gas_midsize

cmb_midsize = pd.DataFrame()
cmb_midsize[['Gas', 'Hybrid']] = phev_midsize['Cmb MPG'].str.split('/', expand=True)
cmb_gas_midsize = cmb_midsize['Gas'].astype(int).mean()
cmb_ph_midsize = cmb_midsize['Hybrid'].astype(int).mean()

cmb_phev_midsize = cmb_ph_midsize / cmb_gas_midsize

phev_midsize_ratios = (city_phev_midsize, hwy_phev_midsize, cmb_phev_midsize)


city_smallsuv = pd.DataFrame()
city_smallsuv[['Gas', 'Hybrid']] = phev_smallsuv['City MPG'].str.split('/', expand=True)
city_gas_smallsuv = city_smallsuv['Gas'].astype(int).mean()
city_ph_smallsuv = city_smallsuv['Hybrid'].astype(int).mean()

city_phev_smallsuv = city_ph_smallsuv / city_gas_smallsuv


hwy_smallsuv = pd.DataFrame()
hwy_smallsuv[['Gas', 'Hybrid']] = phev_smallsuv['Hwy MPG'].str.split('/', expand=True)
hwy_gas_smallsuv = hwy_smallsuv['Gas'].astype(int).mean()
hwy_ph_smallsuv = hwy_smallsuv['Hybrid'].astype(int).mean()

hwy_phev_smallsuv = hwy_ph_smallsuv / hwy_gas_smallsuv

cmb_smallsuv = pd.DataFrame()
cmb_smallsuv[['Gas', 'Hybrid']] = phev_smallsuv['Cmb MPG'].str.split('/', expand=True)
cmb_gas_smallsuv = cmb_smallsuv['Gas'].astype(int).mean()
cmb_ph_smallsuv = cmb_smallsuv['Hybrid'].astype(int).mean()

cmb_phev_smallsuv = cmb_ph_smallsuv / cmb_gas_smallsuv

phev_smallsuv_ratios = (city_phev_smallsuv, hwy_phev_smallsuv, cmb_phev_smallsuv)
```

## Visualing the trends(Midsize)

### Module(s) Used: Matplotlib

Upon compiling the MPG ratios for midsize vehicles it is time to visualise them in a scatter plot. THis will allow us to see the trend of how fuel efficiency varies with degree of electrification.

The preparations for all of these scatter plots have a commonality, making use of the arrays x and y to plot the various scatter diagrams.

The array x is the array of point values for the horizontal axis denoting the degree of electrification access that are plotted. This is calculated as the ratio of the capacities of average batteries in all those fuel types(in kWh) based on the capacity for EVs. Specifically, the average capacities for gasoline-only, hybrid, PHEV and electric vehicles are 0kWh, 1.3kWh, 13.5kWh and 71.7kWh respectively.

The array y meanwhile is the array of point values for the vertical axis denoting the relative fuel efficiency ratios that are plotted.

From all of the plots, we can observe that fuel efficiency increases as the degree of electrifcation rises, but at a decreasing rate. You can even observe how fuel efficiency rises fairly quicky till PHEVs at which point it slows down greatly to reach the EV MPG ratios. This is most pronounced in the highway fuel efficiency plot, possibly since there is minimal regenerative braking the EVs can use to recharge and increase their distance travelled.

```python
x = np.array([0, 1.3, 13.5, 71.7])/71.7
y = [1]
y.extend(list(midsize_ratios['city']))
y = tuple(y)
plt.scatter(x, y)
plt.title("Fuel efficiency against degree of electrification(City, Midsize Cars)")
plt.xlabel("Degree of electrification(0 for gasoline-only, 1 for BEV)")
plt.ylabel("Fuel Efficiency Ratio(MPG(fuel type) per MPG(gasoline))")
plt.savefig("ElectrificationEfficiencyCorrelationMidsizeSedanCity.png")
plt.show()
```

# Conclusion

In this second part of the project, we have managed to visualise and draw observations from the trends of how fuel efficiency varies with the degree of electrification.

We have found out that while intuitively, relative fuel efficiency increases with a higher degree of electrification, the increase is also unexpectedly, at a decreasing rate. This may suggest that there is a case of diminishing returns for fuel efficiency.

At the same time, it seems that this case of diminishing returns may be more apparent in smaller vehicles than larger ones. This is supported by our analysis where there is a slower rate of decrease in the increase in fuel efficiency values for small SUVs than midsize cars in our analysis.

Further examination into this idea with more diverse datasets in the future could help to confirm these findings and accordingly provide key insights on the optimal level of electrification one should look out for in various vehicle classes and use cases to maximise fuel efficiency.
